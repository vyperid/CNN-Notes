```python
import cv2
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline #When using the 'inline' backend, our matplotlib graphs will be included in our notebooks, below our code instead of a pop up screen
sns.set(color_codes=True)#Calling this will change how shorthand codes like “b” or “g” are interpreted by matplotlib in subsequent plots.
```

```python
#Reading the image with imread()
image = cv2.imread('cat.jpeg')#imread() helps in loading an image into jupyter including its pixel values
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)) #opencv loads an image in the format of BGR by default, here we change the format into RGB

plt.show()
```

```python
image.shape #our image is 247x204(pixels) with each pixel having 3 color channels(RGB), 247 is the height, 204 is the weight
```

```python
image[0][0] #RGB color channels of the first pixel
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) #converting image to the greyscale
plt.imshow(cv2.cvtColor(gray, cv2.COLOR_BGR2RGB))#BGR to RGB

plt.show()
```

```python
gray.shape #now our image consist of 1 channel only. Remember greyscale values are between 0-255 0 is Black and 255 is White
```

```python
import numpy as np

data = np.array(gray)
#flatten() returns a copy of the array collapsed into one dimension. For example a[[2,3],[4,5]] to a[2,3,4,5]
flattened = data.flatten() #Now we have the greyscale values of the 50.388 pixels in the array format
flattened.shape 
```
